# DAY 3 — Exploratory Data Analysis (EDA)
## Mutual Fund Analytics | India | 2022–2026
---
**Objective:** Visualise patterns across NAV trends, AUM growth, SIP inflows, investor demographics, geographic distribution, and portfolio holdings using Plotly and Seaborn.


## ⚙️ Setup & Data Loading

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings, os

warnings.filterwarnings('ignore')
os.makedirs('reports', exist_ok=True)

# Seaborn theme
sns.set_theme(style='whitegrid', palette='muted', font_scale=1.1)

# Paths
proc = 'data/processed/'

# Load all datasets
nav         = pd.read_csv(proc + '02_nav_history.csv',          parse_dates=['date'])
aum         = pd.read_csv(proc + '03_aum_by_fund_house.csv',    parse_dates=['date'])
sip         = pd.read_csv(proc + '04_monthly_sip_inflows.csv',  parse_dates=['month'])
cat_inflow  = pd.read_csv(proc + '05_category_inflows.csv',     parse_dates=['month'])
folio       = pd.read_csv(proc + '06_industry_folio_count.csv', parse_dates=['month'])
perf        = pd.read_csv(proc + '07_scheme_performance.csv')
txn         = pd.read_csv(proc + '08_investor_transactions.csv',parse_dates=['transaction_date'])
holdings    = pd.read_csv(proc + '09_portfolio_holdings.csv')
fund_master = pd.read_csv(proc + '01_fund_master.csv')
benchmark   = pd.read_csv(proc + '10_benchmark_indices.csv',    parse_dates=['date'])

print("✅ All datasets loaded!")
for name, df in [('nav', nav), ('aum', aum), ('sip', sip), ('cat_inflow', cat_inflow),
                 ('folio', folio), ('perf', perf), ('txn', txn),
                 ('holdings', holdings), ('fund_master', fund_master)]:
    print(f"   {name:15s}: {df.shape[0]:6,} rows × {df.shape[1]} cols")
```

---
## 📈 Chart 1 — Daily NAV Trend Analysis (2022–2026)
Plot daily NAV for all 40 schemes. Highlight 2023 bull run and 2024 market corrections.


```python
# Merge NAV with fund names
nav_m = nav.merge(fund_master[['amfi_code', 'scheme_name', 'sub_category']], on='amfi_code')

# Select 10 large cap / bluechip funds for clarity
bluechip_codes = [119551, 120503, 118632, 119092, 120841,
                  125497, 118825, 120594, 120503, 118989]
nav_sel = nav_m[nav_m['amfi_code'].isin(bluechip_codes)].copy()
nav_sel['scheme_short'] = nav_sel['scheme_name'].str[:30]

fig = px.line(
    nav_sel, x='date', y='nav', color='scheme_short',
    title='📈 Daily NAV Trend — Bluechip Funds (2022–2026)',
    labels={'nav': 'NAV (₹)', 'date': 'Date', 'scheme_short': 'Fund'},
    template='plotly_white'
)

# Highlight 2023 bull run
fig.add_vrect(x0='2023-01-01', x1='2023-12-31',
              fillcolor='green', opacity=0.08,
              annotation_text='2023 Bull Run',
              annotation_position='top left',
              annotation_font_color='green')

# Highlight 2024 market correction
fig.add_vrect(x0='2024-01-01', x1='2024-06-30',
              fillcolor='red', opacity=0.08,
              annotation_text='2024 Correction',
              annotation_position='top right',
              annotation_font_color='red')

fig.update_layout(
    height=550, legend=dict(orientation='v', x=1.01, font_size=10),
    title_font_size=16, margin=dict(r=220)
)
fig.write_image('reports/chart_01_nav_trend.png', width=1400, height=550, scale=1.5)
fig.show()
print("✅ Saved → reports/chart_01_nav_trend.png")
```

> **🔍 EDA Finding 1:** All bluechip funds showed strong upward NAV trends during 2023, driven by a broad equity market rally. A mild correction was observed in early-mid 2024, followed by recovery. See `chart_01_nav_trend.png`.

---
## 🏦 Chart 2 — AUM Growth by Fund House (2022–2025)
Grouped bar chart showing yearly AUM per fund house. Highlight SBI's dominance.


```python
aum['year'] = aum['date'].dt.year
aum_yr = aum.groupby(['year', 'fund_house'])['aum_lakh_crore'].mean().reset_index()

plt.figure(figsize=(16, 7))
ax = sns.barplot(
    data=aum_yr, x='fund_house', y='aum_lakh_crore',
    hue='year', palette='Blues_d', edgecolor='white'
)

# Highlight SBI bar
for bar in ax.patches:
    if bar.get_height() > 10:
        bar.set_edgecolor('gold')
        bar.set_linewidth(2.5)

plt.title('🏦 AUM Growth by Fund House (2022–2025)', fontsize=15, fontweight='bold', pad=15)
plt.xlabel('Fund House', fontsize=12)
plt.ylabel('Average AUM (Lakh Crore ₹)', fontsize=12)
plt.xticks(rotation=40, ha='right')
plt.legend(title='Year', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.annotate('★ SBI dominates
  at ₹12.5L Cr',
             xy=(0, aum_yr[aum_yr['fund_house'].str.contains('SBI', na=False)]['aum_lakh_crore'].max()),
             fontsize=10, color='darkred', fontweight='bold')
plt.tight_layout()
plt.savefig('reports/chart_02_aum_growth.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved → reports/chart_02_aum_growth.png")
```

> **🔍 EDA Finding 2:** SBI Mutual Fund consistently leads with the highest AUM across all years, reaching approximately ₹12.5 Lakh Crore, followed by HDFC and ICICI. All fund houses showed year-on-year AUM growth, reflecting increasing investor confidence. See `chart_02_aum_growth.png`.

## 💰 Chart 3 — Monthly SIP Inflow Time-Series (Jan 2022 – Dec 2025)
Plotly line chart annotating the ₹31,002 Cr all-time high in Dec 2025.


```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=sip['month'], y=sip['sip_inflow_crore'],
    mode='lines+markers', name='SIP Inflow',
    line=dict(color='#1976D2', width=2.5),
    marker=dict(size=5, color='#1976D2'),
    fill='tozeroy', fillcolor='rgba(25,118,210,0.08)'
))

# Annotate all-time high
max_idx = sip['sip_inflow_crore'].idxmax()
max_month = sip.loc[max_idx, 'month']
max_val   = sip.loc[max_idx, 'sip_inflow_crore']

fig.add_annotation(
    x=max_month, y=max_val,
    text=f'<b>All-Time High<br>₹{max_val:,} Cr</b>',
    showarrow=True, arrowhead=2, arrowcolor='red',
    font=dict(color='red', size=12),
    bgcolor='white', bordercolor='red', borderwidth=1,
    ay=-60
)

fig.update_layout(
    title='💰 Monthly SIP Inflows (Jan 2022 – Dec 2025)',
    xaxis_title='Month', yaxis_title='SIP Inflow (₹ Crore)',
    template='plotly_white', height=480,
    title_font_size=16
)
fig.write_image('reports/chart_03_sip_inflow.png', width=1400, height=480, scale=1.5)
fig.show()
print("✅ Saved → reports/chart_03_sip_inflow.png")
```

> **🔍 EDA Finding 3:** SIP inflows grew steadily from Jan 2022 and hit an all-time high of ₹31,002 Cr in Dec 2025, showing a consistent upward trend. This reflects growing retail investor participation in mutual funds through systematic investing. See `chart_03_sip_inflow.png`.

## 🗺️ Chart 4 — Category Inflow Heatmap (Months × Categories)
Seaborn heatmap with months on X-axis and fund categories on Y-axis.


```python
cat_inflow['month_str'] = cat_inflow['month'].dt.strftime('%b %Y')
pivot = cat_inflow.pivot_table(
    index='category', columns='month_str', values='net_inflow_crore'
)
# Keep month order
month_order = cat_inflow.sort_values('month')['month_str'].unique().tolist()
pivot = pivot.reindex(columns=[m for m in month_order if m in pivot.columns])

plt.figure(figsize=(22, 5))
sns.heatmap(
    pivot, cmap='RdYlGn', center=0,
    linewidths=0.4, linecolor='#e0e0e0',
    cbar_kws={'label': 'Net Inflow (₹ Crore)', 'shrink': 0.8},
    yticklabels=True
)
plt.title('🗺️ Category-wise Net Inflow Heatmap (2022–2025)',
          fontsize=15, fontweight='bold', pad=15)
plt.xlabel('Month', fontsize=11)
plt.ylabel('Fund Category', fontsize=11)
plt.xticks(rotation=75, ha='right', fontsize=8)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.savefig('reports/chart_04_category_heatmap.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved → reports/chart_04_category_heatmap.png")
```

> **🔍 EDA Finding 4:** Equity categories (Large Cap, Flexi Cap, Mid Cap) consistently show higher net inflows (green) while Debt categories occasionally show outflows (red). The heatmap reveals strong seasonal patterns with higher equity inflows in Q4 of each year. See `chart_04_category_heatmap.png`.

## 👥 Chart 5 — Investor Demographics
Age group distribution pie chart, SIP amount box plot by age group, and gender split.


```python
fig, axes = plt.subplots(1, 3, figsize=(20, 7))
colors_age    = ['#42A5F5','#66BB6A','#FFA726','#AB47BC','#EF5350','#26C6DA']
colors_gender = ['#2196F3','#E91E63','#4CAF50']

# Plot 1: Age group pie
age_counts = txn['age_group'].value_counts().sort_index()
axes[0].pie(age_counts.values, labels=age_counts.index,
            autopct='%1.1f%%', startangle=140,
            colors=colors_age, pctdistance=0.82,
            wedgeprops=dict(edgecolor='white', linewidth=1.5))
axes[0].set_title('Age Group Distribution', fontsize=13, fontweight='bold', pad=15)

# Plot 2: SIP box plot by age group
sip_txn = txn[txn['transaction_type'] == 'Sip']
age_order = sorted(sip_txn['age_group'].unique())
sip_by_age = [sip_txn[sip_txn['age_group'] == a]['amount_inr'].values for a in age_order]
bp = axes[1].boxplot(sip_by_age, labels=age_order, patch_artist=True,
                     medianprops=dict(color='red', linewidth=2))
for patch, color in zip(bp['boxes'], colors_age):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1].set_title('SIP Amount by Age Group', fontsize=13, fontweight='bold', pad=15)
axes[1].set_xlabel('Age Group')
axes[1].set_ylabel('SIP Amount (₹)')
axes[1].tick_params(axis='x', rotation=30)

# Plot 3: Gender split pie
gender_counts = txn['gender'].value_counts()
axes[2].pie(gender_counts.values, labels=gender_counts.index,
            autopct='%1.1f%%', startangle=90,
            colors=colors_gender,
            wedgeprops=dict(edgecolor='white', linewidth=1.5))
axes[2].set_title('Gender Split', fontsize=13, fontweight='bold', pad=15)

plt.suptitle('👥 Investor Demographics Analysis', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('reports/chart_05_investor_demographics.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved → reports/chart_05_investor_demographics.png")
```

> **🔍 EDA Finding 5:** The 25–34 age group is the largest investor segment, indicating strong youth participation in mutual funds. SIP amounts tend to be higher in the 35–44 age bracket, suggesting greater financial capacity. Male investors outnumber female investors, highlighting an opportunity to improve financial inclusion. See `chart_05_investor_demographics.png`.

## 🗾 Chart 6 — Geographic Distribution
Horizontal bar chart of SIP amount by state + T30 vs B30 vs T30-adjacent city tier pie.


```python
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Plot 1: SIP by state
state_sip = (txn.groupby('state')['amount_inr']
               .sum()
               .sort_values(ascending=True))
colors_bar = ['#1976D2' if v == state_sip.max() else '#90CAF9' for v in state_sip.values]
axes[0].barh(state_sip.index, state_sip.values / 1e7, color=colors_bar, edgecolor='white')
axes[0].set_title('Total SIP Amount by State', fontsize=13, fontweight='bold', pad=12)
axes[0].set_xlabel('Total Amount (₹ Crore)', fontsize=11)
axes[0].set_ylabel('State', fontsize=11)
axes[0].tick_params(axis='y', labelsize=9)

# Plot 2: City tier pie
tier_counts = txn['city_tier'].value_counts()
colors_tier = ['#FF6384', '#36A2EB', '#FFCE56', '#4BC0C0']
axes[1].pie(tier_counts.values, labels=tier_counts.index,
            autopct='%1.1f%%', startangle=140,
            colors=colors_tier,
            wedgeprops=dict(edgecolor='white', linewidth=1.5),
            pctdistance=0.82)
axes[1].set_title('T30 vs B30 City Tier Distribution', fontsize=13, fontweight='bold', pad=12)

plt.suptitle('🗾 Geographic Distribution of Investments', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('reports/chart_06_geographic.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved → reports/chart_06_geographic.png")
```

> **🔍 EDA Finding 6:** Maharashtra, Delhi and Karnataka lead in total SIP investment amounts, reflecting the concentration of high-income urban investors. T30 cities (top 30 by AUM) dominate investments, but B30 (beyond T30) cities show meaningful participation, indicating growing penetration in tier-2 and tier-3 cities. See `chart_06_geographic.png`.

## 📊 Chart 7 — Industry Folio Count Growth
Line chart from 13.26 Cr (Jan 2022) to 26.12 Cr (Dec 2025) with milestones marked.


```python
fig = go.Figure()

# Total folios
fig.add_trace(go.Scatter(
    x=folio['month'], y=folio['total_folios_crore'],
    mode='lines+markers', name='Total Folios',
    line=dict(color='#1976D2', width=3),
    marker=dict(size=7, color='#1976D2')
))

# Equity folios
fig.add_trace(go.Scatter(
    x=folio['month'], y=folio['equity_folios_crore'],
    mode='lines', name='Equity Folios',
    line=dict(color='#43A047', width=2, dash='dot')
))

# Debt folios
fig.add_trace(go.Scatter(
    x=folio['month'], y=folio['debt_folios_crore'],
    mode='lines', name='Debt Folios',
    line=dict(color='#E53935', width=2, dash='dot')
))

# Milestones
start_val = folio['total_folios_crore'].iloc[0]
end_val   = folio['total_folios_crore'].iloc[-1]
mid_idx   = len(folio) // 2
mid_val   = folio['total_folios_crore'].iloc[mid_idx]
mid_month = folio['month'].iloc[mid_idx]

for m, v, label in [
    (folio['month'].iloc[0],  start_val, f'Start: {start_val:.2f} Cr'),
    (mid_month,               mid_val,   f'Mid: {mid_val:.2f} Cr'),
    (folio['month'].iloc[-1], end_val,   f'End: {end_val:.2f} Cr'),
]:
    fig.add_annotation(x=m, y=v, text=f'<b>{label}</b>',
                       showarrow=True, arrowhead=2,
                       font=dict(size=11, color='#1565C0'),
                       bgcolor='white', bordercolor='#1565C0',
                       borderwidth=1, ay=-45)

fig.update_layout(
    title='📊 Industry Folio Count Growth (2022–2025)',
    xaxis_title='Month', yaxis_title='Folios (Crore)',
    template='plotly_white', height=500,
    title_font_size=16,
    legend=dict(orientation='h', yanchor='bottom', y=1.02)
)
fig.write_image('reports/chart_07_folio_growth.png', width=1400, height=500, scale=1.5)
fig.show()
print("✅ Saved → reports/chart_07_folio_growth.png")
```

> **🔍 EDA Finding 7:** Total industry folios nearly doubled from 13.26 Crore in Jan 2022 to 26.12 Crore in Dec 2025, reflecting massive retail investor participation. Equity folios drove the majority of this growth, while debt folio growth remained relatively flat. See `chart_07_folio_growth.png`.

## 🔗 Chart 8 — NAV Return Correlation Matrix
Pairwise correlation of daily returns for 10 selected funds using Seaborn heatmap.


```python
# Select 10 funds with most NAV history
top10_codes = (nav.groupby('amfi_code')['date']
                  .count()
                  .sort_values(ascending=False)
                  .head(10).index.tolist())

nav_10 = nav[nav['amfi_code'].isin(top10_codes)].copy()
nav_pivot = nav_10.pivot_table(index='date', columns='amfi_code', values='nav')

# Daily returns
returns = nav_pivot.pct_change().dropna()

# Short names for display
code_to_name = fund_master.set_index('amfi_code')['scheme_name'].to_dict()
returns.columns = [code_to_name.get(c, str(c))[:25] for c in returns.columns]

corr = returns.corr()

plt.figure(figsize=(13, 10))
mask = np.triu(np.ones_like(corr, dtype=bool), k=1)
sns.heatmap(
    corr, annot=True, fmt='.2f', cmap='coolwarm',
    center=0, vmin=-1, vmax=1,
    linewidths=0.5, linecolor='white',
    cbar_kws={'label': 'Pearson Correlation', 'shrink': 0.8},
    square=True
)
plt.title('🔗 NAV Return Correlation Matrix — 10 Selected Funds',
          fontsize=14, fontweight='bold', pad=15)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.yticks(rotation=0, fontsize=9)
plt.tight_layout()
plt.savefig('reports/chart_08_correlation_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved → reports/chart_08_correlation_matrix.png")
```

> **🔍 EDA Finding 8:** Funds within the same sub-category (e.g., two Large Cap funds) show high positive correlation (>0.90), suggesting they move together with the market. Funds across categories (e.g., Debt vs Equity) show lower correlation, confirming diversification benefits of holding across categories. See `chart_08_correlation_matrix.png`.

## 🍩 Chart 9 — Sector Allocation Donut Chart
Aggregate sector weights from portfolio_holdings across all equity funds.


```python
sector_wt = (holdings.groupby('sector')['weight_pct']
                     .sum()
                     .sort_values(ascending=False))
total = sector_wt.sum()
sector_pct = (sector_wt / total * 100).round(2)

# Group small sectors into 'Others'
threshold = 3.0
small = sector_pct[sector_pct < threshold]
sector_pct = sector_pct[sector_pct >= threshold]
sector_pct['Others'] = small.sum()
sector_pct = sector_pct.sort_values(ascending=False)

cmap = plt.cm.get_cmap('tab20', len(sector_pct))
colors = [cmap(i) for i in range(len(sector_pct))]

fig, ax = plt.subplots(figsize=(13, 9))
wedges, texts, autotexts = ax.pie(
    sector_pct.values, labels=None,
    autopct=lambda p: f'{p:.1f}%' if p > 4 else '',
    startangle=90, colors=colors,
    pctdistance=0.82,
    wedgeprops=dict(width=0.5, edgecolor='white', linewidth=2)
)
for at in autotexts:
    at.set_fontsize(9)
    at.set_fontweight('bold')

ax.legend(wedges, [f'{s} ({v:.1f}%)' for s, v in zip(sector_pct.index, sector_pct.values)],
          loc='center left', bbox_to_anchor=(1.05, 0.5), fontsize=10)
ax.set_title('🍩 Sector Allocation — Equity Fund Portfolio Holdings',
             fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('reports/chart_09_sector_donut.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Saved → reports/chart_09_sector_donut.png")
```

> **🔍 EDA Finding 9:** Financial Services and IT sectors dominate equity fund holdings, collectively accounting for the largest share of portfolio weights. This aligns with India's index composition and reflects fund manager preference for large, liquid companies in these high-growth sectors. See `chart_09_sector_donut.png`.

## ⚖️ Chart 10 — Risk vs Return Scatter (Bonus Chart)
Sharpe ratio vs 3-year return scatter plot, colour-coded by risk grade.


```python
fig = px.scatter(
    perf, x='std_dev_ann_pct', y='return_3yr_pct',
    color='risk_grade', size='aum_crore',
    hover_name='scheme_name',
    hover_data=['fund_house', 'sharpe_ratio', 'expense_ratio_pct'],
    title='⚖️ Risk vs Return — All 40 Schemes (3-Year)',
    labels={
        'std_dev_ann_pct': 'Annualised Std Dev % (Risk)',
        'return_3yr_pct': '3-Year Return %',
        'risk_grade': 'Risk Grade'
    },
    template='plotly_white', size_max=40
)

# Add quadrant lines
med_risk   = perf['std_dev_ann_pct'].median()
med_return = perf['return_3yr_pct'].median()
fig.add_hline(y=med_return, line_dash='dash', line_color='gray', opacity=0.5)
fig.add_vline(x=med_risk,   line_dash='dash', line_color='gray', opacity=0.5)

fig.add_annotation(x=perf['std_dev_ann_pct'].min()+1, y=perf['return_3yr_pct'].max()-1,
                   text='Low Risk<br>High Return ⭐', font=dict(color='green', size=11))
fig.update_layout(height=550, title_font_size=16)
fig.write_image('reports/chart_10_risk_return.png', width=1400, height=550, scale=1.5)
fig.show()
print("✅ Saved → reports/chart_10_risk_return.png")
```

> **🔍 EDA Finding 10:** Funds in the upper-left quadrant (low risk, high return) represent the best risk-adjusted performers. Large Cap Direct funds tend to cluster here with strong Sharpe ratios. High-risk funds with small AUM in the lower-right quadrant delivered below-median returns, indicating poor risk-adjusted performance. See `chart_10_risk_return.png`.
